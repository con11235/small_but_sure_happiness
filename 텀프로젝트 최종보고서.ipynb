## 1. 주제 선정 이유

### 1) 해커톤 대회 주제
텀프로젝트 전 참여했던 해커톤 대회의 큰 주제인 '대학과 지역상권'으로부터 '경희대학교 주변 지역상권'에 대한 주제를 생각해보게 되었습니다.<br>
평소 공강시간이나 여가시간에 학교 주변에서 '주어진 돈'을 바탕으로 '무엇을' 할수 있을지 고민하는 경우가 많기 때문에 이를 해결해줄 수 있는 프로그램을 만들보고자 하였습니다.<br>
해커톤 대회 내에서 이를 해결해줄 수 있는 프로그램을 만들어보았고, 이 프로그램의 활용도를 높일 수 있는 방안을 고안해보았습니다.<br>
'주어진 돈'으로 학교 주변에서 할 수 있는 활동들을 추천하는 프로그램이었기 때문에 '주어진 돈'을 효율적으로 관리할 수 있으면 더 좋은 프로그램이 될 것이라고 생각되었습니다.<br>
따라서 해당 프로그램을 가계부 형태의 기능을 할 수 있도록 보완하고자 하였고, '경희대학교 주변 상권 데이터 활용 및 사용자 가계부 기반 프로그램'이라는 주제를 선정하게 되었습니다.

### 2) 사용가능한 모듈의 활용
웹파이썬 수업시간에 input이라는 기능을 통해 사용자와 프로그램이 소통할 수 있는 방법을 배우게 된 후, 프로그램과 문자열로만 소통할 수 있다는 것에 부족함을 느꼈습니다.<br>
흔히 게임을 하듯 프로그램과 사용자가 즉각적으로 소통할 수 있는 방법은 없을지에 대해 알아보던 중, 'Pygame'모듈에 대해 알게 되었습니다.<br>
'pygame'은 사용자와의 즉각적인 입력으로서의 반응 말고도, 시각적으로 표현할 수 있는 기능이 다양하기 때문에 분석한 데이터를 시각적으로 보여주기 적합할 것이라 생각하였습니다.<br>
또한 근처 상권 데이터를 직접 모아 csv파일로 만든 후, file을 열어 데이터를 가공하는 방식을 이용했습니다.<br>
따라서 데이터를 분석하고, pygame 모듈을 통해 사용자의 즉각적인 입력과 더불어 시각적인 효과도 주는 프로그램을 만들었습니다.

- 프로그램에서 추천 받은 후 사용한 돈을 입력할 수 있는 창이 새로 나오며, 입력 시 용돈에서 차감되며 그 용돈을 바탕으로 또다시 N등분 해 하루 사용가능한 돈이 입력됩니다.

- 카페의 경우 공부하려고 생각하는 시간을 입력받으며 (ex. 새벽 3시 등) 그 시간에 맞추어 가장 영업시간이 많이 남은 카페 or 도서관을 추천합니다.

- 먹거리, 놀거리, 공부하기 모두 한번에 이용 가능하며, 각각 순차적으로 이용한 후 이용한 요금을 입력해야합니다.
- 먹거리의 경우 메뉴 추천은 계산되어진 하루 사용 가능 용돈에 기반하여 랜덤으로 추천됩니다. (ex. 3000원 -> 0원~3000원 사이의 음식 랜덤 추천)
- 놀거리의 경우 본인의 선택으로 정보를 알 수 있으며 추천되진 않습니다.
- 게임의 모든 내용은 학교 전방 1000m이내 실제 존재하는 상권을 대상으로 합니다.

### 음식 데이터 가공 소스코드

```python
def foodchoice(country,money):
    # foodchoice(country,money)
    # country : '한식','중식','일식','양식' 중 하나를 str의 형태로 입력 받습니다.
    # money : 하루에 사용가능한 돈을 str의 형태로 입력받습니다.
    fileMatrix = []
    newfood=[]
    newchoicelist=[]
    # fileMatrix는 csv파일 내의 자료를 여러개의 list형태로 받는 list입니다.
    # newfood는 fileMatrix 내 각 list의 country가 입력받은 country와 일치하는 list들 만을 모아둔 list입니다.
    # newchoicelist는 newfood 내 각 list의 가격 정보가 입력받은 money보다 작은 list들 만을 모아둔 list입니다.
    import random
    with open('hackathon_menu.csv', 'r') as file:
        for lineContent in file:
            fileMatrix.append(lineContent.strip('\n').split(',')) # 가공한 데이터를 불러와 fileMatrix에 넣는 작업을 합니다.
    for i in range(1,len(fileMatrix)):
        if fileMatrix[i][0]==country:
            newfood.append(fileMatrix[i]) # fileMatrix 내 각 list의 country가 입력받은 country와 일치하는지 확인하여 newfood에 넣는 작업을 합니다.
        else:
            continue
    for i in range(1,len(newfood)):
        if int(newfood[i][3])<int(money):
            newchoicelist.append(newfood[i]) # newfood 내 각 list의 가격 정보가 입력받은 money보다 작은지 확인하여 newchoicelist에 넣는 작업을 합니다.
        else:
            continue
    return random.sample(newchoicelist,5) # newchoicelist에서 5개를 랜덤으로 선택하여 return합니다.
```
